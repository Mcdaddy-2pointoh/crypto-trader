# 0. Imports

```python
import yaml
from exchange.exchange_utils import get_exchange, fetch_data
from portfolio.dip_top import backtest, plot_backtest
from portfolio.mathematical import plot_equity
```

# 1. Load Configs

```python
def load_config(path="../config.yaml"):
    with open(path, "r") as f:
        return yaml.safe_load(f)
```

```python
cfg = load_config()
```

# 2. Setup Exchange

```python
exchange =  get_exchange(config=cfg)
```

# 3. Fetch Data

```python
df = fetch_data(config=cfg, exchange=exchange)
```

# 4. Apply strategy

```python
equity_df, trades_df, final_value, returns = backtest(df, config=cfg)
```

```python
plot_backtest(df, trades_df, equity_df, config=cfg)
```

```python
plot_equity(equity_df)
```

```python
print("Final Portfolio Value:", final_value)
print("Returns (%):", returns)
print(trades_df)
```
